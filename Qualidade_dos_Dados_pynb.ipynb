```python
# Lista das colunas de população estimada
colunas_populacao_estimada = ['POPULAÇÃO ESTIMADA 2020', 'POPULAÇÃO ESTIMADA 2019', 'POPULAÇÃO ESTIMADA 2018',
                              'POPULAÇÃO ESTIMADA 2017', 'POPULAÇÃO ESTIMADA 2016', 'POPULAÇÃO ESTIMADA 2015',
                              'POPULAÇÃO ESTIMADA 2014', 'POPULAÇÃO ESTIMADA 2013', 'POPULAÇÃO ESTIMADA 2012',
                              'POPULAÇÃO ESTIMADA 2011']

# Variável para acompanhar se todos os valores são numéricos e positivos
todos_positivos = True

# Iteração sobre as colunas de população estimada
for coluna in colunas_populacao_estimada:
    # Excluir valores NaN e converter para string
    valores_validos = df[coluna].dropna().astype(str)
    # Verificar se todos os valores na coluna são numéricos e positivos
    if not valores_validos.str.isnumeric().all():
        todos_positivos = False
        print(f"Coluna {coluna}: Nem todos os valores são numéricos.")
    else:
        # Verificar se há algum valor negativo
        valores_negativos = valores_validos[valores_validos.astype(int) < 0]
        if not valores_negativos.empty:
            todos_positivos = False
            print(f"Coluna {coluna}: Há valores negativos.")
        # Verificar se há algum valor que não seja um número
        elif not valores_validos.str.match(r'^\d+$').all():
            todos_positivos = False
            print(f"Coluna {coluna}: Há valores não numéricos.")

# Se todos os valores forem positivos e numéricos, exibir uma mensagem
if todos_positivos:
    print("Todos os valores nas colunas de população estimada são numéricos e positivos.")
```
